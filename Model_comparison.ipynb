Phase 1: Data Loading & Inspection
In this phase, we import the foundational libraries and load both real-world datasets Bank Domain & Heart/Medical Domain into Pandas DataFrames.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# load both datasets
df_bank = pd.read_csv("Bank Customer Churn Prediction.csv")
df_heart = pd.read_csv("synthetic_heart_disease_dataset.csv")

print("Phase 1 Complete: Both datasets loaded successfully!")
print(f"Bank Dataset Shape: {df_bank.shape}")
print(f"Heart Dataset Shape: {df_heart.shape}")
```

Phase 2: Exploratory Data Analysis (EDA) & Visualizations
In this phase, we generate 3 custom plots for each dataset (Total 6 plots) to analyze customer/patient balance, continuous feature distributions, and correlation heatmaps.

```python
# Style plot and set figure size
sns.set_theme(style="whitegrid")
plt.figure(figsize=(18, 5))

# Plot 1: Class Balance 
bank_target = 'churn' if 'churn' in df_bank.columns else 'Churn'
plt.subplot(1, 3, 1)
sns.countplot(x=bank_target, data=df_bank, palette='Set2')
plt.title('Bank Dataset: Customer Status Balance')
plt.xlabel('Status (0 = Stayed, 1 = Left Company)')
plt.ylabel('Total Customers')

# Plot 2: Age Distribution
bank_age = 'age' if 'age' in df_bank.columns else 'Age'
plt.subplot(1, 3, 2)
sns.histplot(data=df_bank, x=bank_age, hue=bank_target, kde=True, multiple='stack', palette='muted')
plt.title('Age Distribution: Stayed vs Left')
plt.xlabel('Customer Age')
plt.ylabel('Count')

# Plot 3: Correlation Heatmap
plt.subplot(1, 3, 3)
numeric_bank = df_bank.select_dtypes(include=[np.number])
sns.heatmap(numeric_bank.corr(), annot=True, cmap='coolwarm', fmt=".2f", cbar=False)
plt.title('Bank Dataset: Feature Correlation')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18, 5))

# Plot 4: Class Balance
heart_target = 'HeartDisease' if 'HeartDisease' in df_heart.columns else (df_heart.columns[-1] if 'target' not in df_heart.columns else 'target')
plt.subplot(1, 3, 1)
sns.countplot(x=heart_target, data=df_heart, palette='Pastel1')
plt.title('Heart Disease: Class Balance')
plt.xlabel('Status (0 = Healthy, 1 = Heart Disease)')
plt.ylabel('Count')

# Plot 5: Age Distribution
heart_age = 'Age' if 'Age' in df_heart.columns else ('age' if 'age' in df_heart.columns else df_heart.select_dtypes(include=[np.number]).columns[0])
plt.subplot(1, 3, 2)
sns.histplot(data=df_heart, x=heart_age, hue=heart_target, kde=True, multiple='stack', palette='Dark2')
plt.title('Age Distribution: Healthy vs Patient')
plt.xlabel('Patient Age')
plt.ylabel('Count')

# Plot 6: Correlation Heatmap
plt.subplot(1, 3, 3)
numeric_heart = df_heart.select_dtypes(include=[np.number])
sns.heatmap(numeric_heart.corr(), annot=True, cmap='viridis', fmt=".2f", cbar=False)
plt.title('Heart Disease: Feature Correlation')

plt.tight_layout()
plt.show()
```
